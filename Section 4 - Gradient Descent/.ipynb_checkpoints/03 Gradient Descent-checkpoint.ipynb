# Notebook Imports and Package

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#make data
x_1 = np.linspace(start=-3, stop=3, num=10)
```

```python
# Plot
plt.plot(x_1, f(x_1))
plt.xlabel('X', fontsize=16)
plt.ylabel('f(X)', fontsize=16)
plt.xlim(-3, 3)
plt.ylim(0, 12)
plt.show()
```
