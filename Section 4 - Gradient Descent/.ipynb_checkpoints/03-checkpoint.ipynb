```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1 = np.linspace(start = -3, stop = 3, num = 10)
```

```python
def f(x):
    return x**2 + x + 1
```

```python
plt.plot(x_1, f(x_1))
plt.xlabel('X')
plt.ylabel('f(x)')
plt.style.use('fivethirtyeight')
```

## Slope & Derivative

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[15, 5])

#Cost function
plt.subplot(1, 2, 1)

plt.title('Cost function')
plt.xlabel('X')
plt.ylabel('f(x)')

plt.plot(x_1, f(x_1), color='blue')

plt.style.use('fivethirtyeight')

#Derivative
plt.subplot(1, 2, 2)

plt.title('And it\'s slope')
plt.xlabel('X')
plt.ylabel('df(x)')

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
```

## Python Loops & Gradient Descent

```python
# Python For Loop
for n in range(5):
    print('baaaaaaa', n)
```

```python
# Python while loop
counter = 0
while counter < 7:
    print('counting...',  counter)
    counter = counter + 1
```
