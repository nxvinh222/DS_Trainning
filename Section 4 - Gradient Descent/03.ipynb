```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_1 = np.linspace(start = -3, stop = 3, num = 10)
```

```python
def f(x):
    return x**2 + x + 1
```

```python
plt.plot(x_1, f(x_1))
plt.xlabel('X')
plt.ylabel('f(x)')
plt.style.use('fivethirtyeight')
```

## Slope & Derivative

```python
def df(x):
    return 2*x + 1
```

```python
plt.figure(figsize=[15, 5])

#Cost function
plt.subplot(1, 2, 1)

plt.title('Cost function')
plt.xlabel('X')
plt.ylabel('f(x)')

plt.plot(x_1, f(x_1), color='blue')

plt.style.use('fivethirtyeight')

#Derivative
plt.subplot(1, 2, 2)

plt.title('And it\'s slope')
plt.xlabel('X')
plt.ylabel('df(x)')

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5)
```

## Python Loops & Gradient Descent

```python
# Python For Loop
for n in range(5):
    print('baaaaaaa', n)
```

```python
# Python while loop
counter = 0
while counter < 7:
    print('counting...',  counter)
    counter = counter + 1
```

```python
# Gradient Descent
new_x = 3
previous_x = 0
step = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(1000):
    previous_x = new_x
    gradient = df(new_x)
    new_x = new_x - gradient * step
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    if abs(previous_x - new_x) < precision:
        print('this loop run:', n,'times')
        break
print(new_x)
print(f(new_x))
```

```python
# Gradient Descent calculation on plot
plt.figure(figsize=[20, 5])

#Cost function
plt.subplot(1, 3, 1)

plt.title('Cost function')
plt.xlabel('X')
plt.ylabel('f(x)')

plt.plot(x_1, f(x_1), color='blue', alpha=0.6)

values = np.array(x_list)
plt.scatter(x_list, f(values), color = 'red', s=100, alpha=0.6)

plt.style.use('fivethirtyeight')

#Derivative
plt.subplot(1, 3, 2)

plt.title('And it\'s slope')
plt.xlabel('X')
plt.ylabel('df(x)')

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)

plt.scatter(x_list, slope_list, color = 'red', s=100, alpha=0.6)

#Derivative(zoom)
plt.subplot(1, 3, 3)

plt.title('And it\'s slope')
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)
plt.xlabel('X')
plt.ylabel('df(x)')

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.8)

plt.scatter(x_list, slope_list, color = 'red', s=500, alpha=0.6)
```
